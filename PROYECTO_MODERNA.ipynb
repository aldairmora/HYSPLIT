```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Datos proporcionados
data = {
    "Meses": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", 
              "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"],
    "Bolivia_2018": [166, 71, 139, 245, 716, 1066, 3730, 5259, 9122, 2754, 824, 427],
    "Brasil_2018": [2553, 1476, 2659, 1656, 3366, 5790, 12652, 22774, 42251, 19668, 13014, 5113],
    "Colombia_2018": [1486, 9047, 3814, 585, 177, 106, 434, 979, 477, 359, 423, 1307],
    "Ecuador_2018": [98, 33, 46, 46, 33, 33, 271, 277, 454, 434, 654, 431],
    "Perú_2018": [202, 158, 309, 59, 177, 317, 969, 267, 2580, 611, 258, 291],
    "AOD_2018": [0.155, 0.126, 0.206, 0.240, 0.348, 0.317, 0.317, 0.267, 0.293, 0.345, 0.293, 0.225]
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Lista de países para el análisis
paises = ["Bolivia_2018", "Brasil_2018", "Colombia_2018", "Ecuador_2018"]

# Crear gráficos
plt.figure(figsize=(14, 10))

for i, pais in enumerate(paises, 1):
    plt.subplot(2, 2, i)
    x = df[pais]
    y = df["AOD_2018"]
    
    # Ajuste de la regresión lineal
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    y_pred = slope * x + intercept
    
    # Gráfico de dispersión y línea de regresión
    plt.scatter(x, y, label=f'Datos {pais}', color='black')
    plt.plot(x, y_pred, color='red', label=f'y = {slope:.2e}x + {intercept:.2f}\n$R^2$ = {r_value**2:.4f}')
    
    # Configuración del gráfico
    plt.title(f'AOD vs Focos de Incendios - {pais.split("_")[0]}')
    plt.xlabel('Focos de Incendios Forestales')
    plt.ylabel('AOD 500nm')
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.show()
```
